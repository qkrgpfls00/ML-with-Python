```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(size):
    # 0부터 5사이에 수를 size만큼 균등히 나누어 생성
    x=np.linspace(0,5,size)
    y=x*np.sin(x**2)+1
    return (x,y)
def sample(size):
    x=np.linspace(0,5,size)
    y=x*np.sin(x**2)+1+np.random.randn(x.size)*0.5
    return(x,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
X, y =sample(1000)
plt.scatter(X,y,s=3,c="black")
plt.show()
```

```python
X.shape, y.shape
```

```python
#1차원->2차원
X = X.reshape(-1,1)
y = y.reshape(-1,1)
X.shape, y.shape
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
#flatten: 2차원->1차원
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), lr.predict(X).flatten())
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
X_poly[:10]
```

```python
lr = LinearRegression()
lr.fit(X_poly,y)
```

```python
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), lr.predict(X_poly).flatten())
plt.show()
```

```python
poly_features = PolynomialFeatures(degree=16)
X_poly = poly_features.fit_transform(X)
X_poly[:3]
```

```python
lr = LinearRegression()
lr.fit(X_poly,y)
f_x, f_y = f(1000)
plt.plot(f_x, f_y)
plt.scatter(X.flatten(), y.flatten(), s=3, c="black")
plt.plot(X.flatten(), lr.predict(X_poly).flatten())
plt.show()
```
